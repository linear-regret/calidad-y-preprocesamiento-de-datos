## **Ejemplo práctico: Relación no lineal entre dos columnas predictoras**
Supón que estás preparando un modelo para predecir si una persona tiene una enfermedad con base en distintas características. Durante el preprocesamiento, decides analizar la relación entre dos variables predictoras del dataset:

- edad (age)
- riesgo_calculado (una variable calculada artificialmente)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import os
```

- La correlación de Pearson será cercana a cero, ya que esta mide solo relaciones lineales.

```python
# Abrir CSV
df = pd.read_csv("edad_riesgo_calculado.csv")
#df.to_csv(csv_path, index=False)

# Calcular correlación de Pearson
correlacion, _ = pearsonr(df['edad'], df['riesgo_calculado'])
print(correlacion)
```

- Sin embargo, al graficar con un scatter plot, se revela una relación en forma de parábola: a edades extremas (jóvenes o mayores), el riesgo calculado es más alto, y es más bajo cerca de los 40.

```python
# Crear scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['edad'], df['riesgo_calculado'], alpha=0.7)
plt.title(f'Relación no lineal entre "edad" y "riesgo_calculado"\nCorrelación de Pearson = {correlacion:.2f}')
plt.xlabel('Edad')
plt.ylabel('Riesgo calculado')
plt.grid(True)

# Guardar imagen
image_path = "edad_vs_riesgo_plot.png"
plt.savefig(image_path)
```

Este ejemplo muestra que:

- Solo basarte en la correlación de Pearson para eliminar o seleccionar variables predictoras puede llevar a descartar variables útiles.

- Es importante visualizar las relaciones entre columnas, especialmente si sospechas de relaciones no lineales.

- Alternativas como la correlación de Spearman (que mide correlación monótona) o técnicas basadas en mutual information pueden capturar relaciones más complejas.
